```python
# import the pandas library
import pandas as pd
```

```python
#Load the data the historical data of stock
from google.colab import files
files.upload()
```

```python
#Read in the data
df = pd.read_csv('TSLA2.csv')
#show the data
df
```

```python
#Select the date and price
df = df[['Date' , 'Close']]

#rename the deatures for prophet
df = df.rename(columns={'Date':"ds", 'Close':'y'})
#show the data
df
```

```python
#Get the last 20 rows of data and store them into a new variable
# use this dataset to predict the value accuracy 
last = df[len(df)-20:]
last
```

```python
#Get all of the rows in the data set except for the last 20 rows of data
df = df[:-20]
df
```

```python
#import facebook library
from fbprophet import Prophet
#Create the Prophet object (model)
fbp = Prophet(daily_seasonality= True)
#fit or train the model
fbp.fit(df)

#predict the future 
future = fbp.make_future_dataframe(periods=365)
forecast = fbp.predict(future)
```

```python
#import the facebook prophet plotting libraries
from fbprophet.plot import plot_plotly
#plot the data
plot_plotly(fbp, forecast)
```

```python
#see how model did prediction with the future price
# use 2021-11-19
forecast[forecast.ds == '2021-11-19']['yhat'] 
```

```python
#show thee actual value for the stock for 2021-11-19
last[last.ds == '2021-11-19']['y']
```
